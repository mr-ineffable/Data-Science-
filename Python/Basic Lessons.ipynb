```python
#Operator Overloading
class Vector:
    def __init__(self,a,b):
        self.x = a
        self.y = b

    def __add__(self, other):
        return Vector(self.x+other.x , self.y+other.y)
    
    def __repr__(self):
        return f"{self.x} + {self.y}i"
    
v1 = Vector(2,3)
v2 = Vector(3,4)

v3 = v1+v2
print(v3)
```

```python
#parameters vs args
#arguments are values passed while calling a function while parameters are defined in function def
#we can give default value to a parameter
def greet(name='Tah'):
    print("Hello", name)
cnt = 10
def fight():
    global cnt
    cnt -= 1
    if(cnt == 0):
        return "Bye"
def actions():
    print("You are stupid Priyanshu")
    var = fight()
    if(var == 'Bye'):
        print("Bye")
        return
    print("You are beautiful")
    if(cnt >0):
        actions()
greet()
actions()
```

```python
#class constructor Python __init__
class Car:
    def __init__(self,name):
        self.name = name

class Specs(Car):
    def __init__(self,name,power):
        super().__init__(name)#if this is not called then name won't exist for child class
        #super() is used to call init of parents class
        self.power = power

    def checkNameAndPower(self):
        return self.name, self.power
    
    def renameName(self,newName):
        self.name = newName

    #renaming can only be done of parent class once using super that variable is initialized


s = Specs("Porshe",500)
print(s.checkNameAndPower())
s.renameName("Ferrari")
print(s.checkNameAndPower())

```
