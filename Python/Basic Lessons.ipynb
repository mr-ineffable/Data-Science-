```python
#Operator Overloading
class Vector:
    def __init__(self,a,b):
        self.x = a
        self.y = b

    def __add__(self, other):
        return Vector(self.x+other.x , self.y+other.y)
    
    def __repr__(self):
        return f"{self.x} + {self.y}i"
    
v1 = Vector(2,3)
v2 = Vector(3,4)

v3 = v1+v2
print(v3)
```

```python
marks = [97,98,97]
#find the index of an element
print(marks.index(97))
#return only the first index of the occurance

#to count of the appearance of an item
print(marks.count(97))

#extending a list

marks.extend([89,88])
print(marks)

#finding minimum and maximum value in a list
print(min(marks) , max(marks))

#iterating a list in a reverse order
for i in range(len(marks)-1, -1,-1):
    print(marks[i], end = ' ')
```

```python
#length of a multi dimensional list
l = [[1,2,3],[4,5,6],[7,8,9]]
print(len(l))

#reversing a list
for i in l[::-1]:
    print(i[::-1])

l.reverse()
for i in l:
    i.reverse()
    print(i)
```

```python
#list comprehension 
#first for loop executes then the given condition is checked 
#if the condition satisfies it is added
lst = [i for i in range(21) if i %2 == 0]
print(lst)
```

```python
multidimensional_list = [[i for i in range(5)] for j in range(10)]
print(multidimensional_list)

#flattening a multidimensional list
flattened_list = [j for i in multidimensional_list for j in i]
print(flattened_list)
```

```python
#merging two dict or updating
dict_1 = {'a':1,'b':2}
dict_2 = {'b':20,'c':3}
merged_1 = dict_1|dict_2
merged_2 = {**dict_1, **dict_2}
dict_1.update(dict_2)

print(dict_1)
print(merged_1)
print(merged_2)
```

```python
#dictionary comprehension
com_dct = {i:i**2 for i in range(1,6)}
print(com_dct)

#dictionary comprehension with condition
con_dct = {i:i**2 for i in range(1,11) if i%2==0}
print(con_dct)
```

```python
#parameters vs args
#arguments are values passed while calling a function while parameters are defined in function def
#we can give default value to a parameter
def greet(name='Tah'):
    print("Hello", name)
cnt = 10
def fight():
    global cnt
    cnt -= 1
    if(cnt == 0):
        return "Bye"
def actions():
    print("You are stupid Priyanshu")
    var = fight()
    if(var == 'Bye'):
        print("Bye")
        return
    print("You are beautiful")
    if(cnt >0):
        actions()
greet()
actions()
```
