```python
#Operator Overloading
class Vector:
    def __init__(self,a,b):
        self.x = a
        self.y = b

    def __add__(self, other):
        return Vector(self.x+other.x , self.y+other.y)
    
    def __repr__(self):
        return f"{self.x} + {self.y}i"
    
v1 = Vector(2,3)
v2 = Vector(3,4)

v3 = v1+v2
print(v3)
```
